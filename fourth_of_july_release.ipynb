Scraping Google Maps to find all lodging businesses within a 1.5 kilometer radius of the Sag Harbor center, in NY.

```python
import pandas as pd
import json
import time
import pprint
import googlemaps
```

```python
# Define API Key
API_KEY = 'your-api-key-here'

# Define our client
gmaps = googlemaps.Client(key= API_KEY)
```

```python
places_nb_results = gmaps.places_nearby(location='40.956463, -72.279838', radius=15000, type='lodging')
```

```python
hamptons_details = pd.DataFrame(data=None, columns=['name', 'formatted_phone_number', 'formatted_address', 'website'])
```

```python
hamptons_details
```

```python
# loop through each place in the results to get price levels with Place Details
for place in hamptons_df['place_id']:
    
#   define my place id
    my_place_id = place
    
#   define the fields we want sent back to us
    my_fields = ['name', 'formatted_phone_number', 'formatted_address', 'website']
    
#   make a request for the details
    place_details = gmaps.place(place_id = my_place_id, fields=my_fields)
    
    #   refined json
    filtered_details = place_details['result']
    
#   add to df
    hamptons_details = hamptons_details.append(filtered_details, ignore_index=True)
```

```python
hamptons_details
```

```python
hamptons_merged = pd.merge(hamptons_df, hamptons_details, on='name', how='left')
```
